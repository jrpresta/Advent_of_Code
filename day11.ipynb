```python
grid_size = 5
grid_serial_num = 1308

def power_calculator(x, y):
    rack_id = x + 10
    power_level = rack_id * y
    power_level += grid_serial_num
    power_level = power_level * rack_id
    power_level = (power_level // 100) % 10
    power_level += -5
    return power_level

grid = [[power_calculator(i,j) for i in range(grid_size)] for j in range(grid_size)]

def grid_sum(x,y):
#     return sum([grid[i][j] for i, j in (range(x-1, x+2), range(y-1,y+2))])
    return sum([sum([grid[i][j] for i in range(x-1, x+2)]) for j in range(y-1, y+2)])
```
